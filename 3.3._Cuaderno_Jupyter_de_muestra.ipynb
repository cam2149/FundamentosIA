# Cuaderno Jupyter de Muestra

### ¿Qué es un cuaderno de Jupyter?

Los cuadernos de Jupyter son una herramienta que permite intercalar celdas de textos y celdas de código Python ejecutable.

Es una herramienta ideal para empezar a programar porque nos permite añadir notas o apuntes así como ir probando el código de manera interactiva a medida que lo vamos elaborando. Además es una herramienta ideal para la docencia por su facilidad de uso e interacción y también por la posibilidad de ser modificados de manera sencilla por los alumnos.

Las dos prácticas que veremos a lo largo de este curso se presentarán en cuadernos de Jupyter que podrán ser modificados y ejecutados por los alumnos.

### ¿Qué funcionalidades tiene un cuaderno de Jupyter?

En primer lugar y como mencionamos observamos que es posible intercalar celdas de texto y celdas de código ejecutable:





```python
print("¡Bienvenido/a a esta clase!")
```

El entorno de ejecución de este código se conserva a lo largo de todo el cuaderno. Esto quiere decir que si definimos una variable en una celda:

```python
saludo = "Hola"
```

Esta variable seguirá existiendo en las celdas posteriores y podrá ser utilizada:

```python
print(saludo)
```

__Nota.__ Es importante tener en cuenta que la ejecución del código del cuaderno es a demanda del usuario, es decir, que si ejecutamos una celda esto no quiere decir que las celdas anteriores hayan sido ejecutadas, lo que puede llevar a errores. Veámoslo con un ejemplo:

```python
despedida = "Hasta la próxima"
```

```python
print(despedida)
```

Por último y como hemos visto en las clases previas una de las funcionalidades más interesantes de Python a la hora de programar es la existencia de bibliotecas de código para realizar funcionalidades específicas.

Veámos un ejemplo de cómo importar y utilizar bibliotecas de Python para realizar una pequeña gráfica:

Para importar las bibliotecas simplemente:

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark")
```

Una vez importados los módulos podemos usar las funciones contenidas en los mismos.

Empezamos generando unos datos de muestra:

```python
muestra = 10000
media = [0, 0]
covarianza = [(2, .4), (.4, .2)]
rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(media, covarianza, muestra).T
```

Tras esto generamos la visualización que encadena distintos gráficos:

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(x=x, y=y, levels=5, color="w", linewidths=1)
```

Con cuatro líneas de código hemos sido capaces de visualizar los datos de una manera bastante compleja sin necesidad de construir funciones complicadas o perder el tiempo en generar todo el marco de la imagen desde cero. Esta es la potencia del uso de módulo que veremos más en profundidad en las próximas clases prácticas.
